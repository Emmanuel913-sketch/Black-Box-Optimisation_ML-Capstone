```python
import numpy as np
import os

BASE_DIR = r"C:\Users\nmnko\Desktop\NEW ML 2025\Module 12\Data\initial_data for functions Mini lesson 12.8"
```

```python

# ================================
# WEEKLY DATA (EDIT ONLY THIS PART)
# ================================

# Each function gets one input vector and one output value
weekly_inputs = {
    1: [0.731022, 0.732999],
    2: [[0.649362, 0.035130],
    3: [0.161259, 0.509609, 1.090670],
    4: [0.414115, 0.414266, 0.357001, 0.427294],
    5: [0.876470, 0.922541, 0.939484, 1.017643],
    6: [0.204871, 0.941871, 0.006522, 0.971655, 0.000001],
    7: [0.000001, 0.236117, 0.399100, 0.339675, 0.339373, 0.673190],
    8: [0.456885, 0.000001, 0.000001, 0.000001, 0.574344, 0.002113, 0.000001, 0.011956],
}

weekly_outputs = {
    1: 7.718434285295828e-16,
    2: 0.36943511733617773,
    3: -0.6837494426161729,
    4: 0.6865956440412302,
    5: 5087.056716113719,
    6: -1.6788784057732298,
    7: 2.5874596903219063,
    8: 9.2616370190124,
}
```

```python

# ================================
# UPDATE LOGIC (DO NOT EDIT)
# ================================

for func_id in range(1, 9):
    func_dir = os.path.join(BASE_DIR, f"function_{func_id}")

    X_path = os.path.join(func_dir, "initial_inputs.npy")
    Y_path = os.path.join(func_dir, "initial_outputs.npy")

    X = np.load(X_path)
    Y = np.load(Y_path)

    X_new = np.array(weekly_inputs[func_id], dtype=np.float64).reshape(1, -1)
    Y_new = np.array([weekly_outputs[func_id]], dtype=np.float64)

    X_updated = np.vstack((X, X_new))
    Y_updated = np.append(Y, Y_new)

    np.save(X_path, X_updated)
    np.save(Y_path, Y_updated)

    print(f"Updated function_{func_id}: "
          f"X shape {X.shape} → {X_updated.shape}, "
          f"Y shape {Y.shape} → {Y_updated.shape}")

print("\n Weekly data successfully appended for all 8 functions.")
```

```python
# ================================
# READ & PRINT ALL DATA
# ================================

print("\n========== FULL DATA (ALL FUNCTIONS) ==========\n")

for func_id in range(1, 9):
    func_dir = os.path.join(BASE_DIR, f"function_{func_id}")

    X = np.load(os.path.join(func_dir, "initial_inputs.npy"))
    Y = np.load(os.path.join(func_dir, "initial_outputs.npy"))

    print(f"\n\n\nFunction {func_id}")
    print("=" * 60)
    print(f"Input dimension: {X.shape[1]}")
    print(f"Total samples  : {X.shape[0]}\n")

    for idx in range(X.shape[0]):
        inputs_str = ", ".join(
            f"x{i+1}={X[idx, i]:.6e}" for i in range(X.shape[1])
        )
        print(f"Sample {idx+1:03d} | {inputs_str} | y={Y[idx]:.6e}")

print("\n\n\n All data printed successfully.")
```
